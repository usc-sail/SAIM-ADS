```python
#checking previous files (previous ACM submission) for topic understanding

train_jwt_ads_of_world_file="/data/digbose92/ads_complete_repo/ads_existing/scripts/ads_cvpr/pkl_files/jwt_ads_of_world_train_final_filtered.pkl"
val_jwt_ads_of_world_file="/data/digbose92/ads_complete_repo/ads_existing/scripts/ads_cvpr/pkl_files/jwt_ads_of_world_valid_final_filtered.pkl"

with open(train_jwt_ads_of_world_file,"rb") as f:
    jwt_ads_of_world_train_data=pickle.load(f)
    
with open(val_jwt_ads_of_world_file,"rb") as f:
    jwt_ads_of_world_val_data=pickle.load(f)
```

```python
def generate_file_names(dict_data):
    
    label_list=[]
    file_list=[]
    for key in list(dict_data.keys()):
        file_list=file_list+dict_data[key]
        label_list=label_list+[key]*len(dict_data[key])
        
    return(label_list,file_list)
        
jwt_ads_of_world_tr_lab_list,jwt_ads_of_world_tr_file_list=generate_file_names(jwt_ads_of_world_train_data)
jwt_ads_of_world_val_lab_list,jwt_ads_of_world_val_file_list=generate_file_names(jwt_ads_of_world_val_data)
```

```python
#extract all jwt files from there

jwt_id_tr_file_list=[i for i,f in enumerate(jwt_ads_of_world_tr_file_list) if 'jwt_ads' in f]
jwt_id_val_file_list=[i for i,f in enumerate(jwt_ads_of_world_val_file_list) if 'jwt_ads' in f]

#jwt file list 
jwt_tr_file_list=[jwt_ads_of_world_tr_file_list[i] for i in jwt_id_tr_file_list]
jwt_val_file_list=[jwt_ads_of_world_val_file_list[i] for i in jwt_id_val_file_list]

#jwt label list
jwt_tr_label_list=[jwt_ads_of_world_tr_lab_list[i] for i in jwt_id_tr_file_list]
jwt_val_label_list=[jwt_ads_of_world_tr_lab_list[i] for i in jwt_id_val_file_list]

jwt_total_file_list=jwt_tr_file_list+jwt_val_file_list
jwt_total_label_list=jwt_tr_label_list+jwt_val_label_list
jwt_file_keys=[os.path.splitext(j.split("/")[-1])[0] for j in jwt_total_file_list]
```

```python
#check if the keys in still_missing present in jwt_file_keys
cvpr_missing_list=[]
recovered_jwt_missing_samples=0
for key in still_missing:
      if(key in jwt_file_keys):
            recovered_jwt_missing_samples+=1
            index_jwt=jwt_file_keys.index(key)
            dict_topic_list[jwt_file_keys[index_jwt]]=jwt_total_label_list[index_jwt]
      else:
        cvpr_missing_list.append(key)
```
