```python
import os 
import pandas as pd 
import numpy as np 
import json
```

```python
json_file="/data/digbose92/codes/ads_codes/SAIM-ADS/data/SAIM_data/language_list.json"
with open(json_file, "r") as f:
    language_list=json.load(f)
```

```python
language_list['toyota_esperas'][0]['label']
```

```python
from collections import Counter
language_names=[language_list[key][0]['label'] for key in list(language_list.keys())]
```

```python
Counter(language_names)
```

```python
language_dictionary_map="/data/digbose92/codes/ads_codes/SAIM-ADS/data/SAIM_data/language_dictionary.json"
with open(language_dictionary_map, "r") as f:
    lang_dict=json.load(f)
```

```python
lang_dict
```

```python
language_names_list=[lang_dict[l] for l in language_names]
```

```python
#create bar plots of top 20 languages 

import matplotlib.pyplot as plt
import seaborn as sns

# Choose a color palette
colors = sns.color_palette('pastel')

#top-k categories
k=20
language_distr=dict(Counter(language_names_list).most_common(20))
categories=list(language_distr.keys())
values=[language_distr[k] for k in categories]

# Create bar plot with border lines
plt.bar(categories, values, edgecolor='black', linewidth=2, color=colors)

# Add labels and title
plt.xlabel('Language')
plt.ylabel('Number of samples')
plt.title('Top-20 languages occuring in SAIM-Ads transcripts')

# Orient x-axis labels vertically
plt.xticks(rotation=90)

#save the language distribution in the dataset
plt.savefig('../plot_figs/language_distribution_transcripts_top_20.png',dpi=300)
```

```python
#obtain score distribution for the individual languages 
#extract topk=20 languages from the dictionary
from tqdm import tqdm 
language_names_code_dict=dict(Counter(language_names).most_common(20))
key_lang=list(language_names_code_dict.keys())

for key_index,v in tqdm(enumerate(key_lang)):
    
    lang_key=key_lang[key_index]
    score_list=[]

    for key in tqdm(list(language_list.keys())):
        if(language_list[key][0]['label']==lang_key):
            score_list.append(language_list[key][0]['score'])

    # Create histogram
    plt.figure(key_index+1)
    sns.distplot(score_list)

    # Set plot title and axis labels
    plt.title('Score Distribution Plot')
    plt.xlabel('Data')
    plt.ylabel('Density')
    fig_folder="/bigdata/digbose92/ads_data/ads_complete_repo/ads_figures"
    fig_filename=fig_folder+"/"+lang_key+".jpg"
    print(fig_filename)
    plt.savefig(fig_filename,dpi=300)
```

```python
#section for whisper language detection outputs 
import os 
import pandas as pd 
import numpy as np 
import json

whisper_json_file="/data/digbose92/ads_complete_repo/ads_codes/SAIM-ADS/data/SAIM_data/language_metadata/language_detection_whisper.json"
with open(whisper_json_file, "r") as f:
    whisper_language_list=json.load(f)
    
#language mapping
lang_map_file="/data/digbose92/ads_complete_repo/ads_codes/SAIM-ADS/data/SAIM_data/language_metadata/FLEURS_id_to_lang.json"
with open(lang_map_file, "r") as f:
    lang_map=json.load(f)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

print(len(lang_map))
language_names=[whisper_language_list[key]['label'] for key in list(whisper_language_list.keys())]
lang_names=[lang_map[id] for id in language_names if id in lang_map]

# Choose a color palette
colors = sns.color_palette('pastel')

#top-k categories
k=20
language_distr=dict(Counter(lang_names).most_common(20))
print(language_distr)
categories=list(language_distr.keys())
values=[language_distr[k] for k in categories]

# Create bar plot with border lines
plt.bar(categories, values, edgecolor='black', linewidth=2, color=colors)

# Add labels and title
plt.xlabel('Language')
plt.ylabel('Number of samples')
plt.title('Top-20 languages occuring in SAIM-Ads transcripts(Whisper language detection)')

# Orient x-axis labels vertically
plt.xticks(rotation=90)
#save the language distribution in the dataset
plt.savefig('../plot_figs/language_distribution_transcripts_top_20_whisper.png',dpi=300)
```
