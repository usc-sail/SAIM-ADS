```python
#augment the current csv file with topic labels 
import pandas as pd
import numpy as np
from tqdm import tqdm

csv_file="/data/digbose92/ads_complete_repo/ads_codes/SAIM-ADS/data/SAIM_data/SAIM_ads_data_message_tone_train_test_val_clip_features_links_augmented.csv"
saim_ads_merged_data=pd.read_csv(csv_file)
print(saim_ads_merged_data.columns)

topic_list=[]
video_file_list=[]
start_tone_list=[]
middle_tone_list=[]
end_tone_list=[]
transition_val_list=[]
social_message_list=[]
split_list=[]
clip_feature_path_list=[]
link_list=[]
video_key_list=[]
for i in tqdm(np.arange(saim_ads_merged_data.shape[0])):
    
    video_key=os.path.splitext(saim_ads_merged_data['video_file'].iloc[i])[0]
    
    
    if(video_key in mapped_topic_list):
        
        video_key_list.append(video_key)
        topic_list.append(mapped_topic_list[video_key])
        #video file list, start, middle and ending tone, tone transition, social message
        video_file_list.append(saim_ads_merged_data['video_file'].iloc[i])
        start_tone_list.append(saim_ads_merged_data['start_tone'].iloc[i])
        middle_tone_list.append(saim_ads_merged_data['middle_tone'].iloc[i])
        end_tone_list.append(saim_ads_merged_data['end_tone'].iloc[i])
        transition_val_list.append(saim_ads_merged_data['Transition_val'].iloc[i])
        social_message_list.append(saim_ads_merged_data['social_message'].iloc[i])
        split_list.append(saim_ads_merged_data['Split'].iloc[i])
        clip_feature_path_list.append(saim_ads_merged_data['clip_feature_path'].iloc[i])
        link_list.append(saim_ads_merged_data['Link'].iloc[i])


df_updated=pd.DataFrame({'video_file':video_file_list,
            'start_tone':start_tone_list,
            'middle_tone':middle_tone_list,
            'end_tone':end_tone_list,
            'Transition_val': transition_val_list,
             'social_message': social_message_list,
             'Topic':topic_list,
             'Split':split_list,
              'clip_feature_path':clip_feature_path_list,
              'Link':link_list})

print(df_updated.shape)
df_updated.to_csv("../data/SAIM_data/SAIM_multi_task_tone_soc_message_data.csv")
```

```python
#create train,test,val split with jwt data in train always 
#70% (5880) in train, 10% (840) in val, 20% (1680) in test 
from random import sample

#read the total file list
video_filename_list="/data/digbose92/ads_complete_repo/ads_codes/SAIM-ADS/data/lomond_available_file_path.txt"
with open(video_filename_list, "r") as f:
    filename_list=f.readlines()
    
video_filename_set=[v.split("\n")[0] for v in filename_list]
jwt_video_filename=[v for v in video_filename_set if 'jwt' in v]#1135
#include 5880 - 1135 = 4745 videos here 
jwt_video_filename_set=[j.split("/")[-1] for j in jwt_video_filename]

def generate_train_test_val_split(total_filename_set,jwt_video_filename_set):
    
    #generate train, test and val list of files
    split_list=['NA']*len(total_filename_set)
    
    #jwt file list 
    for jwt_file in jwt_video_filename_set:
        
        index=total_filename_set.index(jwt_file)
        split_list[index]='train'
        
    #sample 4745 files from the rest of the total file set (5880-4745)
    
    #extract the indices from the rest of the file set where the split value is not train
    id_diff=[i for i,v in enumerate(split_list) if v!='train']
    
    #randomly sample 4745 index values from id_diff
    train_id_rest=sample(id_diff,4745)
    
    for id in train_id_rest:
        split_list[id]='train'
        
    #take difference of train_id_rest with id_diff
    val_test_id_set=list(set(id_diff)-set(train_id_rest)) #2520
    
    #add 840 and 1680 index values 
    val_id=sample(val_test_id,int(0.33*len(val_test_id_set))) #840
    
    #rest all ids would be test
    test_id=list(set(val_test_id_set)-set(val_id))
    
    #validation data
    for id in val_id:
        split_list[id]='val'
        
    #testing data
    for id in test_id:
        split_list[id]='test'
        
    return(split_list)
    
    
    
    
    
    
    
```
