```python
import os 
import pandas as pd 
from videohash import VideoHash
from tqdm import tqdm 
```

```python
video_file_path="/data/digbose92/ads_complete_repo/ads_codes/SAIM-ADS/data/lomond_available_file_path.txt"
with open(video_file_path,'r') as f:
    video_file_list=f.readlines()
```

```python
video_file_list=[v.split("\n")[0] for v in video_file_list]
```

```python
video_hash_list=[]
#video hashing here
for video_file in tqdm(video_file_list):
    
    videohash4 = VideoHash(path=video_file)
    print(videohash4)
    video_hash_list.append(videohash4)
```

```python
#compute average clip embeddings for the entire video
soc_msg_tone_transition_file="/data/digbose92/ads_complete_repo/ads_codes/SAIM-ADS/data/SAIM_ads_data_message_tone_train_test_val_clip_features.csv"
soc_msg_data=pd.read_csv(soc_msg_tone_transition_file)
```

```python
import pickle
import numpy as np
clip_avg_embeddings=np.zeros((soc_msg_data.shape[0],512))

for i in tqdm(np.arange(soc_msg_data.shape[0])):
    clip_file_path=soc_msg_data['clip_feature_path'].iloc[i]
    
    with open(clip_file_path, "rb") as f:
        clip_data=pickle.load(f)
        
    feat=clip_data['Features']
    feat_avg=np.mean(feat,axis=0)
    
    clip_avg_embeddings[i,:]=feat_avg
    
```

```python
clip_avg_embeddings.shape
```

```python
#compute cosine similarity across rows 

def pairwise_cosine_similarity(x):
    """
    x: numpy array of shape (N, D)
    """
    x = x / np.linalg.norm(x, axis=1, keepdims=True)
    return x @ x.T

cosine_sim_values=pairwise_cosine_similarity(clip_avg_embeddings)
```

```python
np.min(cosine_sim_values)
```

```python
#check in the json file similarity of the video links
import json
video_keys=[os.path.splitext(v.split("/")[-1])[0] for v in video_file_list]
json_file="/data/digbose92/ads_complete_repo/ads_of_world_metadata/ads_of_the_world_links_first_version.json"
with open(json_file, 'r') as f:
    ads_of_world_links=json.load(f)
```

```python
dict_download_link={}

for key in list(ads_of_world_links.keys()):
    dict_download_link[key]=ads_of_world_links[key]['download_link']
```

```python
#make a list of links for the video files currently used 
dict_available_links=dict()
link_list=[]
unique_vid_key=[]

for i,vid_key in enumerate(video_keys):
    if(vid_key in dict_download_link):
        if(len(dict_download_link[vid_key])==0):
            dict_available_links[vid_key]=""
        else:
            dict_available_links[vid_key]=dict_download_link[vid_key]
            link_list.append(dict_download_link[vid_key])
    else:
        if("jwt" in video_file_list[i]):
            link_list.append('Not available')
            dict_available_links[vid_key]="Not available"
        else:
            link="https://www.youtube.com/embed/"+vid_key
            dict_available_links[vid_key]=link
            link_list.append(link)
            
#     if(len(dict_download_link[vid_key])==0):
#         dict_available_links[vid_key]=""
#     else:
#         dict_available_links[vid_key]=dict_download_link[vid_key]
```
