```python
import json 
json_file="/bigdata/digbose92/ads_data/ads_complete_repo/ads_of_world_metadata/ads_of_the_world_links_updated.json"
with open(json_file, "r") as f:
    ads_of_world=json.load(f)
```

```python
import pandas as pd
import os 
csv_file="/bigdata/digbose92/ads_data/ads_complete_repo/ads_codes/SAIM-ADS/data/SAIM_data/SAIM_multi_task_tone_soc_message_topic_data_no_zero_files.csv"
SAIM_data=pd.read_csv(csv_file)
clip_feature_path_set=list(SAIM_data['clip_feature_path'])
clip_feature_path_set=[os.path.splitext(c.split("/")[-1])[0] for c in clip_feature_path_set]
```

```python
x=[c.lower() for c in clip_feature_path_set]
```

```python
cnt_files=0
for key in x:
    if(key in ads_of_world):
        cnt_files+=1
```

```python
jwt_json_file="/bigdata/digbose92/ads_data/ads_complete_repo/ads_of_world_metadata/jwt_available_ads_english.json"
with open(jwt_json_file, "r") as f:
    jwt_data=json.load(f)
```

```python
import numpy as np
entry_id_list=[jwt_data[i]['ID'] for i in np.arange(len(jwt_data))]
```

```python
cnt_files=0
key_list_np=[]
dict_total_AOW_JWT=dict()

for key in x:
    if(key in ads_of_world):
        dict_total_AOW_JWT[key]=ads_of_world[key]
        cnt_files+=1
    elif(key in entry_id_list):
        id_list_index=entry_id_list.index(key)
        dict_total_AOW_JWT[key]=jwt_data[id_list_index]
        cnt_files+=1
    else:
        key_list_np.append(key)
```

```python
#save dict_total_AOW_JWT

with open("/bigdata/digbose92/ads_data/ads_complete_repo/ads_of_world_metadata/JWT_AOW_total_data.json","w") as f:
    json.dump(dict_total_AOW_JWT,f,indent=4)
```

```python
#extract country information from the list:
from tqdm import tqdm 
country_list=[]
for key in tqdm(list(dict_total_AOW_JWT.keys())):
    
    if('Tags' in dict_total_AOW_JWT[key]):
        if(len(dict_total_AOW_JWT[key]['Tags'])>0):
            country_list.append(dict_total_AOW_JWT[key]['Tags'][2])
    else:
        country_list.append(dict_total_AOW_JWT[key]['Country'])
```

```python
from collections import Counter
#merge UNITED KINGDOM and United Kingdom into UK
#merge United States and USA into USA
#merge AUSTRALIA and Australia into Australia
#merge CANADA and Canada into Canada
#merge New Zealand and NEW ZEALAND into New Zealand


for i,country in enumerate(country_list):
    if(country=='United States'):
        country_list[i]='USA'
    elif((country=='UNITED KINGDOM') or (country=='United Kingdom')):
        country_list[i]='UK'
    elif((country=='AUSTRALIA')):
        country_list[i]='Australia'
    elif((country=='CANADA')):
        country_list[i]='Canada'
    elif((country=='NEW ZEALAND')):
        country_list[i]='New Zealand'
```

```python
dict_set=dict(Counter(country_list).most_common(100))

#save this list
with open("/bigdata/digbose92/ads_data/ads_complete_repo/ads_codes/SAIM-ADS/data/country_AOW_JWT_distribution.json","w") as f:
    json.dump(dict_set,f,indent=4)
```

```python
#club anything below 30 as other 

condensed_dict=dict()
condensed_dict['other']=0
for key in list(dict_set.keys()):
    if(dict_set[key]>=50):
        condensed_dict[key]=dict_set[key]
    else:
        condensed_dict['other']=condensed_dict['other']+dict_set[key]
```
