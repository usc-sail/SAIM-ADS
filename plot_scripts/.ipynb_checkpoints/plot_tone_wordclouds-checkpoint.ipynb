```python
import os 
import pandas as pd 
import numpy as np 
import json
```

```python
json_file="/proj/digbose92/ads_repo/annotations/ads_completed_annotation_v1/processed_json/MM_AU_dataset.json"
with open(json_file, "r") as f:
    MM_AU_dataset=json.load(f)
```

```python
#create wordclouds for start segment 
from tqdm import tqdm 

key_list=list(MM_AU_dataset.keys())

total_start_tone=[]
total_start_reason_list=[]

total_middle_tone=[]
total_middle_reason_list=[]

total_end_tone=[]
total_end_reason_list=[]

for key in tqdm(key_list):
    
    #start reason + tone 
    start_reason_list=MM_AU_dataset[key]['Start_reason']
    start_tone_list=MM_AU_dataset[key]['Start_tone']
    
    total_start_tone=total_start_tone+start_tone_list
    total_start_reason_list=total_start_reason_list+start_reason_list
    
    #middle reason + tone 
    middle_reason_list=MM_AU_dataset[key]['Middle_reason']
    middle_tone_list=MM_AU_dataset[key]['Middle_tone']
    
    total_middle_tone=total_middle_tone+middle_tone_list
    total_middle_reason_list=total_middle_reason_list+middle_reason_list
    
    #ending reason + tone 
    end_reason_list=MM_AU_dataset[key]['End_reason']
    end_tone_list=MM_AU_dataset[key]['End_tone']
    
    total_end_tone=total_end_tone+end_tone_list
    total_end_reason_list=total_end_reason_list+end_reason_list
    
```

```python
len(total_end_tone),len(total_end_reason_list)
```

```python
######### start word cloud for positive tone #####

start_positive_tone_reason=[]

start_pos_id=[i for i,v in enumerate(total_start_tone) if v=='positive']
reason_list_ids=range(len(total_start_reason_list))

for id in start_pos_id:
    if(id in reason_list_ids):
        start_positive_tone_reason.append(total_start_reason_list[id])
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
plt.rcParams["figure.figsize"] = (15,8)
stopwords=wordcloud.stopwords

stopword_list=['positive','negative','elements','show', 'shows', 'beginning','shown']
for s in stopword_list:
    stopwords.add(s)
wordcloud = WordCloud(width = 1500, height = 600, 
                background_color ='white', stopwords=stopwords,
                min_font_size = 10).generate(" ".join(start_positive_tone_reason))
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```

```python
type(wordcloud.stopwords)
```

```python
######### start word cloud for negative tone #####

start_negative_tone_reason=[]
start_neg_id=[i for i,v in enumerate(total_start_tone) if v=='negative']
reason_list_ids=range(len(total_start_reason_list))

for id in start_neg_id:
    if(id in reason_list_ids):
        #print(total_start_reason_list[id])
        start_negative_tone_reason.append(total_start_reason_list[id])
        
plt.rcParams["figure.figsize"] = (15,8)
stopwords=wordcloud.stopwords

wordcloud = WordCloud(width = 1500, height = 600, 
                background_color ='white', stopwords=stopwords,
                min_font_size = 10).generate(" ".join(start_negative_tone_reason))
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```

```python
mid_negative_tone_reason=[]
mid_neg_id=[i for i,v in enumerate(total_middle_tone) if v=='negative']
reason_list_ids=range(len(total_middle_reason_list))

for id in start_neg_id:
    if(id in reason_list_ids):
        mid_negative_tone_reason.append(total_middle_reason_list[id])
        
plt.rcParams["figure.figsize"] = (15,8)
stopwords=wordcloud.stopwords
stopword_list=['positive','elements','show', 'shows', 'beginning','shown','enjoying','happy','smiling', 'dancing','girl','kid', 'people', 'woman','singing','man','visual','car']
for s in stopword_list:
    stopwords.add(s)
mid_negative_tone_reason=[m.lower() for m in mid_negative_tone_reason]

mid_neg_word_list=[]
for sent in mid_negative_tone_reason:
    mid_neg_word_list=mid_neg_word_list+sent.split(" ")

mid_dict_word=Counter(mid_neg_word_list)

wordcloud = WordCloud(width = 1500, height = 600, 
                background_color ='white',
                    stopwords=stopwords,
                min_font_size = 10).generate_from_frequencies(mid_dict_word)
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```
